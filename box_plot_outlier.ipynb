```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.datasets import load_diabetes
iris = load_diabetes() 

df = pd.DataFrame(iris.data, columns =iris.feature_names ) 
import seaborn as sns 
import warnings 
warnings.filterwarnings('ignore')
```

```python
sns.distplot(df['age'])
```

```python
sns.boxplot(df['s3'])
```

```python
# find the iqr range : 
df['s3'].describe()
```

```python
per_25 = df['s3'].quantile(0.25) * 100 
per_75 = df['s3'].quantile(0.75 ) * 100 

# iqr is 
iqr = per_75 - per_25 
iqr
```

```python
# iqr is said : 
low_min = per_25 - 1.5* iqr 
upp_max = per_75 + 1.5 * iqr 
```

```python
df[df['s3'] < low_min]
df[df['s3'] > upp_max] 
```
