# learning curve for visulize the overfitting and under fitting : 

```python
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_iris 
from sklearn.model_selection import cross_val_score,train_test_split 
from sklearn.model_selection import KFold 
from sklearn.metrics import mean_squared_error , r2_score 
import pandas as pd
import numpy as np
from sklearn.model_selection import learning_curve,LearningCurveDisplay
```

```python
x,y = load_iris().data , load_iris().target

x_train,x_test, y_train,y_test = train_test_split(
                x,y,test_size = 0.33 ,random_state = 42 ) 
reg = LinearRegression () 
reg.fit(x_train,y_train)
```

```python
cross_val = KFold(10) 
acc  = cross_val_score (reg,x_train,y_train,cv = cross_val)
acc_train = acc.mean ()
```

```python
y_pred = reg.predict(x_test)
train_score = reg.score(x_train,y_train)
print('traing score : ',train_score , 'crossvalidation ("testing") score',acc_train )
```

```python
lcd = LearningCurveDisplay.from_estimator ( reg, x_train,y_train,score_type = 'both') 
lcd.plot()
```
